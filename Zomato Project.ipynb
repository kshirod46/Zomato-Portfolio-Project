# Data Cleaning

```python
# removing duplicate data
print("Number of duplicates values: ",df1.duplicated().sum())
df1.drop_duplicates(inplace=True)
print("Number of duplicates values: ",df1.duplicated().sum())
print(df1.shape)
```

1. Rate column

```python
df1['rate'].unique()
```

```python
# removing nan,NEW
df1['rate']=df1['rate'].replace(['NEW'],'0')
df1['rate'].unique()
```

```python
# only taking rating out of 5 
df1['rate'] = df1['rate'].str.split('/5').str[0].str.strip()
df1['rate'].dtype
df1
```

```python
# chnaging datatype of rate to float
df1["rate"] = pd.to_numeric(df1["rate"], errors="coerce")
df1["rate"].dtype
```

```python
#no of nan value in rate column 
print("No of Null rating: ",df1['rate'].isnull().sum())
```

```python
# filling nan with 0
df1['rate'] = df1['rate'].fillna(0)
df1
```

 2.making name clear

```python
df1['name'] = df1['name'].str.split(' - ').str[0].str.strip() # delimiter ‘-’ and keeps only the first part (before the delimiter).
df1['name'] = df1['name'].str.split(' - ').str[0].str.strip() # delimiter ‘-’ and keeps only the first part (before the delimiter).
df1['name'] = df1['name'].str.lower() # converting to lowercase as its easier for comparison when working with text data.
df1['name'].unique()
```

```python
df1.tail()
```

```python
# changing all columns name
df1 = df1.rename(columns={"name":"restaurant_name",
                          "approx_cost(for two people)":"cost_for_two",
                          "listed_in(type)":"type",
                          "listed_in(city)":"city",
                          "book_table":"reservations",
                          })
df1
```

3. rest_value

```python
df1['rest_type'].unique()
# categorige the rest_type
print("Grouping by rest_type and count:",df1['rest_type'].value_counts())
```

```python
# we recognize only the count greater  than 500 of each rest type and lesser will be categorised as 'others'
rest_type = df1["rest_type"].value_counts(ascending = False)
rest_less_than_500 = rest_type[rest_type < 500]
def rest_type(data):
    if data in rest_less_than_500:
        return "Other"
    else:
        return data
df1["rest_type"] = df1["rest_type"].apply(rest_type)
df1
```

```python
# checking null values in rests
df1['rest_type'].isnull().sum()
# replacing null with Other
df1['rest_type'].fillna("Other", inplace=True)
df1
```

4. cost_for_two 

```python
df1['cost_for_two'].unique()
# we can see there are  comma , in between some values so first we remove that
#replacing comma with ''
df1['cost_for_two']=df1['cost_for_two'].str.replace(",",'')
df1['cost_for_two'].unique()
```

```python
# chaniging datatype of cost_for_two to float
df1["cost_for_two"]=pd.to_numeric(df1["cost_for_two"],errors="coerce")
df1["cost_for_two"].dtype
```

```python
#no of nan value in cost_for_two column 
print("No of Null rating: ",df1['cost_for_two'].isnull().sum())
print(df1['cost_for_two'].describe())
```

```python
print('mean:',df1['cost_for_two'].mean())
print('median:',df1['cost_for_two'].median())
print('mode:',df1['cost_for_two'].mode())
```

```python
# so after seeing mean,median,mode ,minimum ,standard deviation we cantjust put the null value 0
# so we will assume these null value as median value
df1['cost_for_two'].fillna(df1['cost_for_two'].median(), inplace = True)
df1['cost_for_two'].isnull().sum()
```

# Data Visulalization

```python
# online order possible or not
#the percentage of the online orders in the restauants
colors = [ 'lightblue' , 'lightgreen']
df1['online_order'].value_counts().plot(kind='pie', autopct="%0.1f%%", explode=[0, 0.1], shadow=True,colors=colors)
plt.title('Restaurants delivering Online or not')
plt.ylabel('Online Order Percentage')
```

```python
# table booking or reservations facility avaliable or not
#the percentage of the booking a table in the restauants
colors = [ 'lightblue' , 'lightgreen']
df1['reservations'].value_counts().plot(kind='pie', autopct="%0.1f%%", explode=[0, 0.2], shadow=True, colors =colors)
plt.title('Restaurants allowing table booking or not')
plt.ylabel('book percentage')
```

```python
# finding best location 
#Calculate count of restaurants in each city
city_counts = df1['city'].value_counts().reset_index()
city_counts.columns = ['city', 'count']


#Create our bar plot
sns.barplot(y='city', x='count', data=city_counts, width =0.8 )


plt.xlabel('Restaurant City')
plt.ylabel('Count of restaurants')
plt.title('Number of Restaurants in city')

plt.show()
```

```python
type= sns.countplot(x="type",data=df1)
type.bar_label(type.containers[0], fontsize=10)

plt.title("Types of Services - Number of Sales Point")
plt.xlabel("Types of Services")
plt.ylabel("Number of Sales Points")

plt.show()
```

```python
plt.figure(figsize=(13,10)) 
y = df1["restaurant_name"].value_counts()[:10]
x = y.index
ax = sns.barplot(x=y,y=x,data=df1,estimator="sum", errorbar=None)

ax.bar_label(ax.containers[0], fontsize=10)
plt.title("Top 10 Restaurants - Number of Sales Point")
plt.ylabel("Top 10 Restaurant Names")
plt.xlabel("Number of Sales Points")
plt.show()
```

```python
#get the top 10 restaurant types
top_rest_types = df1['rest_type'].value_counts().head(10)

# Create our count plot to know which type is most common
sns.countplot(x='rest_type', data= df1, order = top_rest_types.index )
plt.xticks(rotation = 90)
```

```python
loc_cost = df1['cost_for_two'].groupby(df1['city'],sort=True)
dict_1={}
for i,j in df1['city'].value_counts()[:10].to_dict().items():
    dict_1[i]=round(loc_cost.get_group(i).mean(),2)
cost_df = pd.DataFrame(list(dict_1.items()),columns=['city',"cost_for_two"])
cols = ['#405DE6', '#5B51D8', '#833AB4', '#C13584', '#E1306C',
        '#FD1D1D', '#F56040', '#777737', '#FCAF45', '#FFDC80']


fig , ax = plt.subplots(figsize=(10,8))
sns.barplot(data = cost_df.sort_values(by=['cost_for_two'],ascending=False),
            x = 'cost_for_two',y = 'city',palette=cols)
plt.title('Top 10 Location - Average cost(2 people)')
```
